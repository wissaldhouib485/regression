```python
import pandas as pd

file_path = '/content/calendar.csv.zip'
calendar_data = pd.read_csv(file_path)


print(calendar_data.info())
```

# Suppression des symboles et conversion en type float

```python
calendar_data['price'] = (
    calendar_data['price'].str.replace('[\$,]', '', regex=True).astype(float)
)
calendar_data['adjusted_price'] = (
    calendar_data['adjusted_price'].str.replace('[\$,]', '', regex=True).astype(float)
)
```

 # Conversion de la colonne date en format datetime

```python
calendar_data['date'] = pd.to_datetime(calendar_data['date'], errors='coerce')
```

#  Conversion des colonnes minimum_nights et maximum_nights en entiers

```python
calendar_data['minimum_nights'] = calendar_data['minimum_nights'].fillna(0).astype(int)
calendar_data['maximum_nights'] = calendar_data['maximum_nights'].fillna(0).astype(int)
```

 # Nettoyage des colonnes booléennes

# Conversion de la colonne available en booléen (True/False)


```python
calendar_data['available'] = calendar_data['available'].map({'t': True, 'f': False})
```
